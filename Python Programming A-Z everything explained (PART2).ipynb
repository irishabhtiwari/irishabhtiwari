## `Creating graph from a file`

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = []
y = []
```

```python
readfile = open("mytestfile","r")
```

```python
data = readfile.read().split("\n")
```

```python
for i in data:
    val = i.split(",")
    x.append(int(val[0]))
    y.append(int(val[1]))
    
print(data)

print(x)
print(y)


plt.plot(x,y)
plt.show()
```

```python
import matplotlib 
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

rec = fig.patch
rec.set_facecolor('white')


x = [3,7,8,12]
y=[5,13,2,8]


graph1 = fig.add_subplot(1,1,1)
graph1.plot (x,y,'red',linewidth =2.0)

graph1.tick_params(axis="x",color = "black")
graph1.tick_params(axis="y",color = "black")


graph1.spines['top'].set_color('green')
graph1.spines['left'].set_color('blue')
graph1.spines['right'].set_color('blue')

graph1.set_title("Amount of rain in september",color="black")
graph1.set_xlabel("No of day rain",color="black")
graph1.set_ylabel("Inches of rain",color="black")


plt.show()
```
